#  載入資料與基本資訊

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

# Basic info
print(penguins.head())
penguins.info()
```

# 結果與分析

# 鰭長與體重的關係
利用點散圖觀察鰭長（Flipper Length）跟體重（Body Mass）的關係


```python
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=penguins)
plt.title("Flipper Length vs Body Mass by Species")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.show()
```

**分析結果：**不同物種間，鰭長越長，體重也越重

# 島嶼分布

```python
sns.countplot(x="island", hue="species", data=penguins)
plt.title("Penguin Species Distribution by Island")
plt.xlabel("Island")
plt.ylabel("Count")
plt.show()
```

**分析結果：**Adelie 是唯一同時分布於三個島嶼（Biscoe、Dream、Torgersen）的物種，數量最多。不同物種有其特定棲地偏好，可能與食物來源或繁殖條件有關。

# 性別與體重

```python
sns.boxplot(x="sex", y="body_mass_g", data=penguins)
plt.title("Body Mass by Sex")
plt.xlabel("Sex")
plt.ylabel("Body Mass (g)")
plt.show()
```

**分析結果：**雄性企鵝的體重普遍高於雌性，且分布範圍也較寬。雌性的體重分布集中在較小區間。

# 體重／喙長比例

```python
penguins["mass_to_bill"] = penguins["body_mass_g"] / penguins["bill_length_mm"]

sns.violinplot(x="species", y="mass_to_bill", data=penguins)
plt.title("Body Mass to Bill Length Ratio by Species")
plt.xlabel("Species")
plt.ylabel("Mass/Bill Length Ratio")
plt.show()
```

**分析結果：**Gentoo 的體重／喙長比例最高且分布較狹窄，顯示其體型壯碩、喙較短。Chinstrap 與 Adelie 的比例較低，尤其 Adelie 的比值分布最廣，代表個體差異大。



# 結論與建議

# 結論
1.企鵝的鰭長與體重之間呈現正相關，體型大的企鵝通常鰭也較長，特別是在 Gentoo 物種中最為明顯。

2.三個島嶼的物種分布具明顯差異，Adelie 為分布最廣的物種，而 Gentoo 與 Chinstrap 則集中於特定島嶼。

3.雄性企鵝的體重普遍高於雌性，且體重分布較廣，顯示性別間存在顯著體型差異。

4.體重與喙長的比例在不同物種之間有顯著差異，反映其生態位與進化適應的差異。
