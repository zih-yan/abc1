```python
import pandas as pd
url = 'https://raw.githubusercontent.com/imchihchao/aop113b/main/materials/04-car_models.csv'
df = pd.read_csv(url)
df
```

```python
df_std = df.drop(columns='名稱')
df_mms = df.drop(columns='名稱')
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
df_std[  df_std.columns.to_list()  ] = std.fit_transform(df_std[  df_std.columns.to_list() ] )
df_std
```

```python
from  sklearn.preprocessing  import  MinMaxScaler
mms = MinMaxScaler()
df_mms[  df_mms.columns.to_list()  ] = mms.fit_transform(df_mms[  df_mms.columns.to_list() ] )
df_mms
```
